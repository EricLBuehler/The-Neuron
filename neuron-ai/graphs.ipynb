```python
with open("train_acc.txt", "r") as file:
    train_acc=[float(item.split("_")[1]) for item in file.read().strip().split("\n")]

with open("test_acc.txt", "r") as file:
    test_acc=[float(item.split("_")[1]) for item in file.read().strip().split("\n")]

with open("train_loss.txt", "r") as file:
    train_loss=[float(item.split("_")[1]) for item in file.read().strip().split("\n")]

with open("test_loss.txt", "r") as file:
    test_loss=[float(item.split("_")[1]) for item in file.read().strip().split("\n")]
```

```python
import matplotlib.pyplot as plt

plt.plot(range(len(train_acc)),train_acc, label="Train")
plt.plot(range(len(train_acc)),test_acc, label="Test")
plt.ylabel("Accuracy %")
plt.xlabel("Epoch")
plt.title("Accuracy graph")
leg = plt.legend(loc='upper left')
plt.show()

plt.plot(range(len(train_loss)),train_loss, label="Train")
plt.plot(range(len(train_loss)),test_loss, label="Test")
plt.ylabel("Loss")
plt.xlabel("Epoch")
plt.title("Loss graph")
leg = plt.legend(loc='upper right')
plt.show()
```
